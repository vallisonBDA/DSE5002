```python
opioid_disp_rts=pd.read_csv("County Opioid Dispensing Rates.csv")
```

```python
sdoh_2019=pd.read_csv("2019_SDOH.csv")
```

```python
sdoh_2020=pd.read_csv("2020_SDOH.csv")
```

```python
#view datasets
sdoh_2019.head(10)
```

```python
#combine sdoh datasets
sdoh_combined=pd.concat([sdoh_2019, sdoh_2020], ignore_index=True)
```

```python
# Compare column names
cols_2019 = set(sdoh_2019.columns)
cols_2020 = set(sdoh_2020.columns)

# Columns that are only in one dataset
only_in_2019 = cols_2019 - cols_2020
only_in_2020 = cols_2020 - cols_2019

print("Columns only in 2019 dataset:", only_in_2019)
print("Columns only in 2020 dataset:", only_in_2020)

# If columns differ, show a summary
if only_in_2019 or only_in_2020:
    print("\n⚠️ Some columns are not shared between datasets.")
else:
    print("\n✅ Column names match perfectly!")

# Optionally, inspect side-by-side for confirmation
print("\n2019 columns:\n", sorted(sdoh_2019.columns))
print("\n2020 columns:\n", sorted(sdoh_2020.columns))
```

```python
want = [
    "ACS_TOT_POP_WT","ACS_AVG_HH_SIZE","ACS_PCT_DISABLE","ACS_PCT_FEMALE","ACS_PCT_NON_CITIZEN",
    "ACS_MEDIAN_AGE","ACS_PCT_AGE_ABOVE65","ACS_PCT_BLACK","ACS_PCT_HISPANIC","ACS_PCT_WHITE",
    "ACS_PCT_CHILD_1FAM","ACS_GINI_INDEX","ACS_MEDIAN_HH_INC","ACS_PCT_HH_FOOD_STMP",
    "ACS_PCT_NO_WORK_NO_SCHL_16_19","ACS_PCT_LT_HS","ACS_MEDIAN_HOME_VALUE","ACS_PCT_MEDICAID_ANY",
    "ACS_PCT_MEDICARE_ONLY","ACS_PCT_PRIVATE_ANY","ACS_PCT_PRIVATE_EMPL","ACS_PCT_TRICARE_VA",
    "ACS_PCT_UNINSURED","AHRF_VET","AHRF_UNEMPLOYED_RATE","AHRF_PCT_PRESC_PEN",
    "AMFAR_TOT_MEDSAFAC","AMFAR_TOT_MHFAC","CEN_POPDENSITY_COUNTY","MP_MEDICARE_ELIGIBLES",
    "NCHS_URCS_2013","SAIPE_PCT_POV","POS_PCT_HOSP_FOR_PROFIT"
]

p2019 = pd.read_csv('/content/2019_SDOH.csv')
p2020 = pd.read_csv('/content/2020_SDOH.csv')

missing_2019 = [c for c in want if c not in p2019.columns]
missing_2020 = [c for c in want if c not in p2020.columns]

print("Missing in 2019:", missing_2019)
print("Missing in 2020:", missing_2020)
```

```python
# Columns to keep
keep_cols = [
    "YEAR", "COUNTYFIPS",
    "ACS_TOT_POP_WT","ACS_AVG_HH_SIZE","ACS_PCT_DISABLE","ACS_PCT_FEMALE",
    "ACS_PCT_NON_CITIZEN","ACS_MEDIAN_AGE","ACS_PCT_AGE_ABOVE65","ACS_PCT_BLACK",
    "ACS_PCT_HISPANIC","ACS_PCT_WHITE","ACS_PCT_CHILD_1FAM","ACS_GINI_INDEX",
    "ACS_MEDIAN_HH_INC","ACS_PCT_HH_FOOD_STMP","ACS_PCT_NO_WORK_NO_SCHL_16_19",
    "ACS_PCT_LT_HS","ACS_MEDIAN_HOME_VALUE","ACS_PCT_MEDICAID_ANY",
    "ACS_PCT_MEDICARE_ONLY","ACS_PCT_PRIVATE_ANY","ACS_PCT_PRIVATE_EMPL",
    "ACS_PCT_TRICARE_VA","ACS_PCT_UNINSURED","AHRF_VET","AHRF_UNEMPLOYED_RATE",
    "AHRF_PCT_PRESC_PEN","AMFAR_TOT_MEDSAFAC","AMFAR_TOT_MHFAC",
    "CEN_POPDENSITY_COUNTY","MP_MEDICARE_ELIGIBLES","NCHS_URCS_2013",
    "SAIPE_PCT_POV","POS_PCT_HOSP_FOR_PROFIT"
]

# Subset to keep only those columns that actually exist in the file
keep_cols_existing = [c for c in keep_cols if c in sdoh_combined.columns]

# Create trimmed dataset
sdoh_trimmed = sdoh_combined[keep_cols_existing].copy()
```

```python
#view trimmed dataset
sdoh_trimmed.head(10)
```

```python
#trim opioid dataset to 2019 and 2020
opioid_disp_rts['YEAR'] = opioid_disp_rts['YEAR'].astype(int)
opioid_trimmed = opioid_disp_rts[opioid_disp_rts['YEAR'].isin([2019, 2020])].copy()
```

```python
#view dattset
opioid_trimmed.head(10)
```

```python
#Remove Data Unavailable rows
# --- Convert opioid_dispensing_rate to numeric, coercing invalid values to NaN ---
opioid_trimmed['opioid_dispensing_rate'] = pd.to_numeric(
    opioid_trimmed['opioid_dispensing_rate'], errors='coerce'
)

# --- Drop rows where the rate is missing (was 'Data unavailable') ---
before = len(opioid_trimmed)
opioid_trimmed = opioid_trimmed.dropna(subset=['opioid_dispensing_rate'])
after = len(opioid_trimmed)
```

```python
#view filtered dataset
opioid_trimmed.head(10)
```

```python
#view shape of opioid dataset trimmed
opioid_trimmed.shape
```

```python
#rename county fips code column to match sdoh dataset
opioid_trimmed.rename(columns={'STATE_COUNTY_FIP_U': 'COUNTYFIPS'}, inplace=True)


```

```python
#make both county fips and year integers in both datasets
opioid_trimmed['YEAR'] = opioid_trimmed['YEAR'].astype(int)
opioid_trimmed['COUNTYFIPS'] = opioid_trimmed['COUNTYFIPS'].astype(int)
sdoh_trimmed['YEAR'] = sdoh_trimmed['YEAR'].astype(int)
sdoh_trimmed['COUNTYFIPS'] = sdoh_trimmed['COUNTYFIPS'].astype(int)


```

```python
#join datasets on year and county fips
merged_data = pd.merge(
    opioid_trimmed,
    sdoh_trimmed,
    on=['COUNTYFIPS', 'YEAR'],
    how='left'
)
```

```python
#view merged data
merged_data.head(10)
```

```python
#view merged data shape
merged_data.shape
```

```python
merged_data.to_csv('merged_opioid_sdoh_data.csv', index=False)
```
